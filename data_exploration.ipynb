```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "JNJ", "JPM", "V", "PG", "XOM", "CVX"]
df = yf.download(tickers, start="2010-01-01", end="2011-01-01", group_by='ticker')

print(df.head())
```

|   Ticker  | Nom complet de l’entreprise | Secteur principal                                 |
| :-------: | :-------------------------- | :------------------------------------------------ |
|  **AAPL** | Apple Inc.                  | Technologie (électronique & logiciels)            |
|  **MSFT** | Microsoft Corporation       | Technologie (logiciels, cloud, IA)                |
| **GOOGL** | Alphabet Inc. (Google)      | Technologie (internet, publicité, cloud)          |
|  **AMZN** | Amazon.com Inc.             | Consommation discrétionnaire / e-commerce / cloud |
|  **JNJ**  | Johnson & Johnson           | Santé / produits pharmaceutiques                  |
|  **JPM**  | JPMorgan Chase & Co.        | Finance / banque                                  |
|   **V**   | Visa Inc.                   | Services financiers / paiements                   |
|   **PG**  | Procter & Gamble Co.        | Biens de consommation (produits ménagers & soins) |
|  **XOM**  | Exxon Mobil Corporation     | Énergie / pétrole & gaz                           |
|  **CVX**  | Chevron Corporation         | Énergie / pétrole & gaz                           |


```python
sample = df['GOOGL']
sample = sample.sort_values('Date')

sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 5))
sns.lineplot(data=sample, x="Date", y="Close", color="tab:blue")
plt.title("Évolution de l'action GOOGLE")
plt.xlabel("Date")
plt.ylabel("Prix de clôture")
plt.show()


sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 5))
for element in tickers:
    sample = df[element]
    sample = sample.sort_values('Date')
    sns.lineplot(x=df.index, y=df[element]['Close'], label=element)
plt.title("Évolution des actions")
plt.xlabel("Date")
plt.ylabel("Prix de clôture")
plt.show()
```
